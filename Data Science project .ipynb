```python
X = df.drop(columns=['high_risk_applicant']).values
y = df['high_risk_applicant'].values
#divide the data into two part x and y -> x is Independent variable -> y is Dependent variable/target variable 
```

```python
from sklearn.model_selection import train_test_split #split  dataset into train and test 80% for train and 20% test
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#It contain lots of categorical veriable so I need to  import LabelEncoder and apply it on all the cetagorical veriables
from sklearn.preprocessing import LabelEncoder
labelencoder_X=LabelEncoder()
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.fit_transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
DTClassifier=DecisionTreeClassifier(criterion='entropy',random_state=0)
DTClassifier.fit(X_train,y_train)
y_pred=DTClassifier.predict(X_test)
from sklearn import metrics
print("the accuracy of decision tree is: ", metrics.accuracy_score(y_pred,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier #Best fit for accuracy
from sklearn.metrics import accuracy_score
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.naive_bayes import GaussianNB
NBClassifier=GaussianNB()
NBClassifier.fit(X_train,y_train)
y_pred = NBClassifier.predict(X_test)
accuracy_score(y_test,y_pred)
```
